## Since the transits method is by far the most popular method for discovering planets, we had to look at it closely.
# Now we consider the transits method.

the probability $P = \frac{R_p + R_s}{a}$.

$P$ is the probability, $R_p$ is the radius of planet, $R_s$ is the radius of star, $a$ is the semi major axis.

we could find the radius of star and radius of planet, semi-major axis from the data.

we could find the min probability based on these data, and we could ues it as the base line for transits method.

Here we uesd the mean value of smallest 5 probability of datafile as the baseline

```python
#drop incomplate data
date_for_tran = Tr_data[["pl_rade","st_rad","pl_orbsmax"]].dropna()
date_for_tran["t_r"] = np.array(date_for_tran["pl_rade"]*u.R_earth) + np.array(date_for_tran["st_rad"]*u.Rsun.to(u.R_earth))
date_for_tran["p"]=np.array(date_for_tran["t_r"]) / (np.array(date_for_tran["pl_orbsmax"]))
min_group = np.array(date_for_tran["p"])
min_group.sort()
min_p_trans = np.mean(min_group[:15]) #
```

The relation between them: $a\cdot p = R_s + R_p$

```python
plt.figure(figsize=(12, 12))
plt.plot(date_for_tran["pl_orbsmax"], date_for_tran["t_r"], ".",color = '#1f77b4',markersize=10,label= 'Transit')
plt.plot(solar_planet_semimajor, solar_planet_radius_e + (1 * u.R_sun).to(u.R_earth), "*",color = 'black',label ='Data from Solar system')
plt.plot(10.0**(np.arange(6)-2) * u.AU,10.0**(np.arange(6)-2) * u.AU*min_p_trans+min(date_for_tran["t_r"])* u.AU, "-",color = '#1f77b4',label ='Sensitivity Limit of Transit',linewidth=3)
plt.plot(solar_planet_semimajor[2], solar_planet_radius_e[2] + (1 * u.R_sun).to(u.R_earth), "*",color = 'red',label ='Target')

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Semi-major axis [au]")
plt.ylabel(r"Radius [R$_{Earth}$]")
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.ylim(10,10**(3))
plt.xlim(10**(-3),10**(2))
plt.legend(prop={'size': 15},loc=4)
```

# II, Calculating Detection Signal of Different Methods

Target planet: Semi-major: 1 AU, Temp:288K Radius:1 R_earth, , Mass: 1 M_earth


### 1, Radial Velocity

$K = \frac{M_p}{M_s}\sqrt{\frac{GM_s}{a}}\sin{i}$

Here set: $i =90$ degree, $\sin{i}=1$ $R_s$ is the radius of star, $R_p$ is the radius of planet, a is semi-major axis, $G$ is the gravitional constant.

More K, more chance of detection

```python
K = ((1*c.R_earth/(1*c.R_sun))*(c.G*c.M_sun/(1*c.au))**(1/2))
print(K)
```

### 2, Radial Velocity
we have two equation here,

1, $P = \frac{R_s+R_p}{a}$, here p is the probability, $R_s$ is the radius of star, $R_p$ is the radius of planet, a is semi-major axis

2, $f=(\frac{R_p}{R_s})^2$

The larger the $P$ and $f$, the higher the chance of detection

```python
P = (c.R_sun+c.R_earth )/c.au
f = (c.R_earth/c.R_sun)**2
print(round(P.value*100,3),"%")
print(round(f.value*10**6,3),"ppm")
```

### 3, Direct Imaging

three equation

1, $f_r = \frac{\pi R_p^2}{4\cdot \pi \cdot a^2}A$, $f_r$ is the Reflected starlight. planet radius $R_p$, orbital radius $a$, albedo $A$

2, $f_c = (\frac{R_p}{R_s})^2 \frac{\exp({h \nu/k_BT_s})-1}{\exp({h \nu/k_BT_p})-1}$, n = frequency, h = Plancks constant, kB = Boltzmann constant

3, $\theta = 1.22\frac{\lambda}{D}$, telescope of diameter D, working at 
wavelength $\lambda$



[albedo](https://energyeducation.ca/encyclopedia/Albedo#:~:text=The%20overall%20albedo%20of%20the,thus%20changes%20Earth's%20heat%20balance.) =0.3

Here we expore the wavelength: from 0.5 - 20 $\mu$m

$Hz =\frac{c}{\lambda}$, c=speed of light, $\lambda$=wavelength.

The telescope: [E-ELT(Extremely Large Telescope)](https://en.wikipedia.org/wiki/Extremely_Large_Telescope), Diameter	39.3 m

```python
wl_um = np.linspace(0.5,20,200)*u.um
fc_um = (c.c/wl_um).to(u.Hz)
```
